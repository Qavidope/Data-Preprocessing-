```python
missing_value_per=df.isnull().sum() / df.shape[0]*100
missing_value_per
```

```python
missing_value_per.shape
```

```python
missing_value_clm_gre_20 = missing_value_per[missing_value_per>20].keys()
missing_value_clm_gre_20
```

```python
df2_drop_clm=df.drop(columns=missing_value_clm_gre_20)
df2_drop_clm.shape   #droping the column which contains missing values 
```

```python
df2_drop_clm.head()
```

```python
df3_num  =  df2_drop_clm.select_dtypes(include=['int64','float64'])
```

```python
df3_num.head()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]    #show which has NaN in it 
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var=  [ var for var in df3_num.columns 
                      if df3_num[var].isnull().sum()>0
                    ]
missing_num_var    # these coulumn name contain NaN  
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate (missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(  df3_num[var]  , bins=20, kde_kws={'linewidth':3,'color':'red'})
```

```python
df4_num_mean=df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
#let compare distplot  of old (df3) and  new (df4)

plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate (missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':3,'color':'red'},label="original")
    sns.distplot(df4_num_mean[var],bins=20,kde_kws={'linewidth':2 , 'color':'green'},label="mean")
    plt.legend()
```

```python
df5_num_median=df3_num.fillna(df3_num.mean())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,hist=False,kde_kws={'linewidth':7,'color':'red'} ,label='original')
    sns.distplot(df4_num_mean[var],bins=20,hist=False,kde_kws={'linewidth':4,'color':'green'} ,label='mean')
    sns.distplot(df5_num_median[var],bins=20,hist=False,kde_kws={'linewidth':1,'color':'k'} ,label='median')
    plt.legend()
   
```

```python
for i,var in enumerate(missing_num_var):
    plt.figure(figsize=(7,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df5_num_median[var])
```

```python
df_concat=pd.concat([df3_num[missing_num_var],
                     df4_num_mean[missing_num_var]
                    ,df5_num_median[missing_num_var]],axis=1)
```
